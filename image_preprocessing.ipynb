```python
!pip install mtcnn
import cv2
import numpy as np  
from mtcnn import MTCNN
import os

# Ініціалізація MTCNN детектора
detector = MTCNN()

def is_frontal_face(keypoints):
    eye_left = keypoints['left_eye']
    eye_right = keypoints['right_eye']
    eye_diff = abs(eye_left[1] - eye_right[1])
    nose = keypoints['nose']
    symmetry = abs(eye_left[0] - nose[0]) - abs(eye_right[0] - nose[0])
    return eye_diff < 5 and abs(symmetry) < 10

def align_face(image, eye_left, eye_right):
    dY = eye_right[1] - eye_left[1]
    dX = eye_right[0] - eye_left[0]
    angle = np.degrees(np.arctan2(dY, dX))
    eyesCenter = ((eye_left[0] + eye_right[0]) // 2, (eye_left[1] + eye_right[1]) // 2)
    M = cv2.getRotationMatrix2D(eyesCenter, angle, scale=1)
    aligned = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]), flags=cv2.INTER_CUBIC)
    return aligned

def crop_and_align_face(image_path):
    image = cv2.imread(image_path)
    results = detector.detect_faces(image)
    if results:
        keypoints = results[0]['keypoints']
        if is_frontal_face(keypoints):
            eye_left = keypoints['left_eye']
            eye_right = keypoints['right_eye']
            aligned_image = align_face(image, eye_left, eye_right)
            results_aligned = detector.detect_faces(aligned_image)
            if results_aligned:
                x, y, width, height = results_aligned[0]['box']
                cropped_face = aligned_image[y:y+height, x:x+width]
                return cropped_face, True
    return None, False

# Перебираємо кожну папку та кожне зображення
root_dir = '/kaggle/input/wiki-crop'  # Потрібно змінити на ваш шлях
output_dir = '/kaggle/working/cropped-faces'  # Потрібно змінити на ваш шлях
os.makedirs(output_dir, exist_ok=True)

# Лічильник для відстеження кількості збережених облич
face_count = 0
target_count = 9000  # Заданий обсяг оброблених зображень

for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        if face_count >= target_count:
            break
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(subdir, file)
            face, found = crop_and_align_face(image_path)
            if found:
                face_count += 1
                output_path = os.path.join(output_dir, f'face_{face_count}.png')
                cv2.imwrite(output_path, face)
                
    if face_count >= target_count:
        break
        
print(f'Total faces cropped: {face_count}')
```

```python
import os
import zipfile

# Папка, яку потрібно перетворити в ZIP
folder_to_zip = '/kaggle/working/cropped-faces'

# Назва ZIP файлу
zip_filename = 'cropped_faces.zip'

# Створення ZIP архіву
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    # Перебираємо усі файли та підкаталоги в папці
    for root, dirs, files in os.walk(folder_to_zip):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, os.path.relpath(file_path, folder_to_zip))

print(f'ZIP file created: {zip_filename}')
```

```python
from facenet_pytorch import InceptionResnetV1
from torchvision import transforms
from PIL import Image
import numpy as np
from sklearn.cluster import DBSCAN
import os
import shutil
from collections import defaultdict

# Initialize InceptionResnetV1 for generating embeddings
resnet = InceptionResnetV1(pretrained='vggface2').eval()

# Define the transformation: resize, convert to tensor, and normalize
transform = transforms.Compose([
    transforms.Resize((160, 160)),  # Resize the image to 160x160 pixels
    transforms.ToTensor(),  # Convert the image to a tensor
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])  # Normalize the image
])

images_folder = '/kaggle/input/face-test'  # Adjust this to your folder path
images_dir = os.listdir(images_folder)[:100]  # Adjust if you want to process a different number of images

embeddings = []
processed_image_paths = []

# Vectorize faces
for img_name in images_dir:
    img_path = os.path.join(images_folder, img_name)
    image = Image.open(img_path).convert('RGB')  # Ensure image is in RGB

    # Apply transformation
    img_transformed = transform(image)

    # Add batch dimension and convert image to embedding
    embedding = resnet(img_transformed.unsqueeze(0)).detach().numpy()  # Unsqueeze to add batch dimension
    embeddings.append(embedding.flatten())
    processed_image_paths.append(img_path)

# Кластеризація за допомогою DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # eps та min_samples потрібно підібрати
clusters = dbscan.fit_predict(np.array(embeddings))

# Групування зображень за кластерами
clustered_images = defaultdict(list)
for image_path, cluster_id in zip(processed_image_paths, clusters):
    clustered_images[cluster_id].append(image_path)

# Функція для очищення директорії
def clear_directory(directory_path):
    if os.path.exists(directory_path):
        shutil.rmtree(directory_path)
    os.makedirs(directory_path, exist_ok=True)
    
# Функція для збереження кластерів як окремих папок
def save_clusters_as_folders(clustered_images, output_dir_base):
    clear_directory(output_dir_base)  # Очищуємо базову директорію перед збереженням нових кластерів
    for cluster_id, images in clustered_images.items():
        cluster_dir = os.path.join(output_dir_base, str(cluster_id))
        os.makedirs(cluster_dir, exist_ok=True)
        for image_path in images:
            shutil.copy(image_path, cluster_dir)

save_clusters_as_folders(clustered_images, '/kaggle/working/clustered_images')
```
